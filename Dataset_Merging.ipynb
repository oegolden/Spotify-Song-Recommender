```python

# import required module
import os
data = json.load(open('10k MillionPlaylist entry\mpd.slice.1000-1999.json'))
playlists = pd.DataFrame(data["playlists"])
directory = '10k MillionPlaylist entry'
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    if os.path.isfile(f):
        playlists = pd.concat([playlists,pd.DataFrame(json.load(open(f))['playlists'])])
```

Merge the playlists data with the million songs data by matching the track id's or the song names and artist names, then use the playlists as training data

```python
tracks = pd.DataFrame(playlists["tracks"])
track_frame = pd.DataFrame(tracks.iloc[0][0])
```

```python

for x in range(1,playlists.__len__()):
    track_frame  = track_frame.append(pd.DataFrame(tracks.iloc[x][0]))

track_frame = track_frame.drop(columns = "pos")

track_frame
```

```python
playlist_list = []
track_frame = track_frame.dropna()
for x in range(0,len(playlists.index)):
    for y in range(0,playlists.iloc[x]["num_tracks"]):
        playlist_list.append(playlists.iloc[x][0])
```

```python
track_frame.reset_index(drop = True)
track_frame["playlist_name"] = playlist_list
track_frame
```

```python
frame_created = False
for x in range(0, 200000):
    li = [track_frame.iloc[x]["track_name"]]
    df_temp = pd.DataFrame(df.iloc[np.where(df.track_name.isin(li))])
    if df_temp.empty:
        continue
    else:
        if track_frame.iloc[x]["artist_name"] not in df['artist_name'].values:
            continue
        else:
            df_new = pd.DataFrame(df_temp.loc[df_temp['artist_name'] == track_frame.iloc[x]["artist_name"]])
            playlist = [track_frame.iloc[x]["playlist_name"]]
            df_new["playlist_name"] = playlist
            if not frame_created:
                trframedata = df_new.copy(deep = True)
                frame_created = True

            else:
                trframedata = pd.concat([trframedata,df_new])
```

```python
trframedata = trframedata.dropna()
trframedata.to_csv('Final_playlist_dataset')
trframedata
```
